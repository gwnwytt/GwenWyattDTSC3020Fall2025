## 0) Setup

```python
# Run it first.
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

sns.set_theme(style='whitegrid')
penguins = sns.load_dataset('penguins')
penguins = penguins.dropna()
penguins.head()  # preview
```

## 1–3) Scatter plot with color by species
Create a scatter plot with:
- **x =** `bill_length_mm`
- **y =** `bill_depth_mm`
- Color points by **species**

Then add a short comment (2–3 sentences) describing the pattern(s) you see.

```python
# TODO: Create the scatter plot here
sns.scatterplot(data=penguins, x='bill_length_mm', y='bill_depth_mm', hue='species')
plt.title("Comparing Bill Length by Depth")
plt.xlabel("Bill Length (mm)")
plt.ylabel("Bill Depth (mm)")
```

**Your interpretation (2–3 sentences):**

_Write your observations here._

The Adelie species majority has a bill length between 35mm and 45mm, and depth between 16mm and 20mm. They have the smallest length of the 3 species with the other two starting at over 40mm.

## 4) Histogram with two different bin sizes
Plot a histogram of `bill_length_mm` **twice**:
- Once with **10 bins**
- Once with **30 bins**

Then write 1–2 sentences comparing what changes when you change the number of bins.

```python
# TODO: Histogram with 10 bins
sns.histplot(data=penguins, x='bill_length_mm', stat='density', bins=10)
plt.title("Histogram of Bill Density (10 bins)")
plt.xlabel("Bill Length (mm)")
```

```python
# TODO: Histogram with 30 bins
sns.histplot(data=penguins, x='bill_length_mm', stat='density', bins=30)
plt.title("Histogram of Bill Density (30 bins)")
plt.xlabel("Bill Length (mm)")
```

**Your comparison (1–2 sentences):**

_Write your observations here._

Upping the amount of bins allows the user to see a lot more detail. In the 10 bin histogram, you would not be able to observe the large dip made in the middle of the 30 bin histogram.

## 5) Boxplot of body mass by species
Create a **boxplot** showing `body_mass_g` grouped by `species`.
Then write 1–2 sentences summarizing which species are heavier/lighter on average.

```python
# TODO: Boxplot of body mass by species
sns.boxplot(data=penguins, x='species', y='body_mass_g')
plt.title("Body Mass(g) by Species")
plt.xlabel('Species')
plt.ylabel('Body Mass (g)')
```

**Your summary (1–2 sentences):**

_Write your observations here._

The Gentoo species has a over 1000g increase in average than the other 2 species.
